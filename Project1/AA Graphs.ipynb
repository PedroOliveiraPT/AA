# Graph Generation
This is a notebook purely rdedicated for the graph generation with the data generated from running the algorithms

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
ops = pd.read_csv("/content/drive/MyDrive/op_count.csv")
times = pd.read_csv("/content/drive/MyDrive/times.csv")
```

```
times.head()
```

```
    def make_graphs(file, y_label):
  df = pd.DataFrame(file)
  plt.figure(figsize=(15,5))
  rec_data = df['Recursive'][:12]
  mem_data = df['Memoization'][:12]
  dp_data = df['Dynamic'][:12]

  # show increasing data first
  plt.subplot(1,3,1)
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.plot(rec_data)

  plt.subplot(1,3,2)
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.plot(mem_data)

  plt.subplot(1,3,3)
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.plot(dp_data)

  plt.tight_layout()
  plt.show()

  # check the log scale
  plt.figure(figsize=(15,5))
  plt.subplot(1,3,1)
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.plot(rec_data)
  plt.yscale('log')

  plt.subplot(1,3,2)
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.plot(mem_data)
  plt.yscale('log')

  plt.subplot(1,3,3)
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.plot(dp_data)
  plt.yscale('log')
  
  plt.tight_layout()
  plt.show()
```

```
make_graphs(ops, "Operation Count")
```

```
make_graphs(times, "Execution Time")
```

```
def make_compare_graphs(file, y_label):
  x = [1, 2, 5, 10, 20, 50, 100, 200, 500, 1000]
  df = pd.DataFrame(file, index=[1, 2, 5, 10] + list(range(11, 17)))
  plt.figure(figsize=(5,5))
  mem_data = df['Memoization']
  dp_data = df['Dynamic']

  # show increasing data first
  plt.plot(x, mem_data, 'bo-', label="Memoization")
  plt.plot(x, dp_data, 'ro--', label="Dynamic")
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.xscale('log')
  plt.legend()
  plt.show()

  # check the log scale
  plt.figure(figsize=(5,5))
  # show increasing data first
  plt.plot(x, mem_data, 'bo-', label="Memoization")
  plt.plot(x, dp_data, 'ro--', label="Dynamic")
  plt.xlabel("String length")
  plt.ylabel(y_label)
  plt.xscale('log')
  plt.yscale('log')
  plt.legend()
  plt.show()
```

```
make_compare_graphs(ops, "Operation Count")
```

```
make_compare_graphs(times)
```
